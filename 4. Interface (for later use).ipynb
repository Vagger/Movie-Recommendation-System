```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import random
```

```python
mov = pickle.load(open('mov.pkl','rb'))
mov.head()
```

```python
release_year = mov.release_date.apply(lambda x: int(x[:4])).rename("release_year")
```

```python
mov = mov.join(release_year)
```

```python
mov_most_budget = mov.sort_values(by=['budget'], ascending=False)
mov_most_popular = mov.sort_values(by=['popularity'], ascending=False)
mov_highest_rated = mov.sort_values(by=['vote_average'], ascending=False)
mov_most_voted = mov.sort_values(by=['vote_count'], ascending=False)
new_movies = mov_most_popular[mov_most_popular.release_year >= 2010]
old_movies = mov_most_popular[mov_most_popular.release_year <= 2000]
old_new_movies = mov_most_popular[(mov_most_popular.release_year > 2000) & (mov_most_popular.release_year <2010)]
```

```python
new_movies.head()
```

```python
def movies_displayed():
    mov1 = mov_highest_rated.head(100).sample(1)
    mov2 = mov_most_popular.head(100).sample(1)
    mov3 = mov_most_budget.head(100).sample(1)
    mov4 = mov_most_voted.head(100).sample(1)
    mov5 = new_movies.head(100).sample(1)
    mov6 = old_movies.head(100).sample(1)
    mov7 = old_new_movies.head(100).sample(1)
    
    displayed = [[mov1.original_title.values[0], mov1.id.values[0]], [mov2.original_title.values[0], mov2.id.values[0]], [mov3.original_title.values[0], mov3.id.values[0]], [mov4.original_title.values[0], mov4.id.values[0]], [mov5.original_title.values[0], mov5.id.values[0]], [mov6.original_title.values[0], mov6.id.values[0]], [mov7.original_title.values[0], mov7.id.values[0]]]
    
    return displayed
```
